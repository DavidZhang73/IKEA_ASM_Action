```python
import matplotlib.pyplot as plt
from i3d import videotransforms
from IKEAActionDataset import IKEAActionVideoClipDataset as Dataset
from torchvision import transforms
import numpy as np
```

```python
train_transform = transforms.Compose([videotransforms.RandomCrop(224), videotransforms.RandomHorizontalFlip()])

# train_transform = transforms.Compose([videotransforms.RandomHorizontalFlip()])

dataset = Dataset(
    dataset_path=r"D:\dataset\ikea_action_dataset_frame_small",
    annotation_path=r"D:\dataset\ikea_action_dataset_video",
    transform=train_transform,
    index_filename="train_dataset_index.txt",
    frames_per_clip=16,
    frame_skip=2,
)
```
